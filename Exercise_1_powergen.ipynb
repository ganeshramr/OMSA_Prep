```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
day = np.linspace(0, 60, 1000) # start, finish, n points
```

```python
m1G = 1 * day
m2G = 4 * (day - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Energy Genration')
plt.xlabel('time (in days)')
plt.ylabel('power gen (in kj)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 100])
ax.plot(day, m1G, c='green')
ax.plot(day, m2G, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
